**Q.1 Create a simple page with some div tags and show different ways to add CSS as well as what happens**
**when you target the same elements with inline, internal, and external CSS. Also, utilize comments in the project where required. **

```python
<!DOCTYPE html>
<html>
<head>
    <meta charset='utf-8'>
    <meta http-equiv='X-UA-Compatible' content='IE=edge'>
    <title>Page Title</title>
    <meta name='viewport' content='width=device-width, initial-scale=1'>
    <link rel="stylesheet" href="style_6.css">
    <style>
        .heading{
            color:red;
        }

        </style>
</head>
<body>
    <div class="heading"> This is an internal css HII!!</div><!--internal css-->
    <div style="color:blueviolet; font: size 22px;"> Welcome to our website of CSS </div><!--inline css-->
    <div class="hello"> This is an external css!!</div><!--external css-->
    <hr>
    <h1>Explanations</h1>
    <p>1. INLINE CSS</p>
    <p style="font-weight: bold;">Applied directly within the head section of the html within the style tags. It affects the entire page.</p>
        <p>2. INternal CSS</p>
    <p style="font-weight: bold;">Applied directly within the style tags of the body section.It applies to heading section. .</p>
        <p>3. EXTERNAL CSS</p>
    <p style="font-weight: bold;">Applied through the link tag in the <head> . The style.css file contains css rules thta can be applied to any html pages linkedd to it.</p>
    <br>
    <h1>What happens when applied together</h1>
    <p style="font-weight: bold;">Overriding: The order of precedence for css styles is : Inline>Internal>external. Therefore, if an element has inline styles and internal styles defined, the inline styles will take precedence, and the internal style will be ignored.</p>
</body>
</html>
```

**Q.2 Build an HTML page with multiple paragraphs, each assigned a unique class name. Write CSS rules using
class selectors to apply distinct styling to each paragraph. Follow the BEM naming convention and
explain how you've named the classes=**

```python
<!DOCTYPE html>
<html>
<head>
    <meta charset='utf-8'>
    <meta http-equiv='X-UA-Compatible' content='IE=edge'>
    <title>Page Title</title>
    <meta name='viewport' content='width=device-width, initial-scale=1'>
    <style>
/* BEM Block */
.paragraph {
  font-family: sans-serif;
  line-height: 1.5;
  margin-bottom: 1em; /* Consistent spacing */
}

/* BEM Block Element */
.paragraph__header {
  font-weight: bold;
  margin-bottom: 0.5em;
}

/* BEM Block Modifier */
.paragraph--important {
  color: red;
}

/* BEM Block Modifier (Another example) */
.paragraph--notice {
  font-style: italic;
}
</style>

</head>
<body>

<!-- Paragraph 1 -->
<p class="paragraph">This is the first paragraph.  It's a general paragraph.</p>

<!-- Paragraph 2 -->
<p class="paragraph paragraph--important">
<span class="paragraph__header">Important Notice:</span> This paragraph is styled as important and has a header.
</p>

<!-- Paragraph 3 -->
<p class="paragraph paragraph--notice">This paragraph is marked as a notice, using italic font.</p>
<hr>
<h1>Explanation:</h1>
<h2>1. BEM Structure:</h2>
<ol>
    <li>paragraph (Block): The base element for the paragraph.</li>
    <li>paragraph__header (Block Element): A child element within the paragraph, specifically the header.</li>
    <li>paragraph--important, paragraph--notice (Block Modifiers): Modify the styles of the base paragraph.</li>
    </ol>
<h2>2. CSS Styling:</h2>
<ol>
    <li>.paragraph: Styles the general paragraph.</li>
    <li>.paragraph__header: Styles the header within the paragraph.</li>
    <li>.paragraph--important: Styles the paragraph as important.</li>
    <li>.paragraph--notice: Styles the paragraph as a notice.</li>
    </ol>
<h2>3. HTML Usage:</h2>
<ol>
    <li>Each paragraph is assigned the base .paragraph class.</li>
    <li>Additional class names (paragraph--important, paragraph--notice, paragraph__header) are added to modify the styles as needed.</li>
    </ol>
<h2>4. BEM Benefits:</h2>
<ol>
    <li>Clarity: The BEM structure makes it easy to understand how different styles are applied and related to each other.</li>
    <li>Maintainability: Changes can be made to the base block or individual modifiers/elements without affecting unrelated parts of the code.</li>
    </ol>


</body>
</html>
```

**Q.3 Develop an HTML form with various input elements. Use CSS to style the form, including setting
background colors for input fields. Create a custom color palette for the form elements, and demonstrate
how to apply opacity to one of the form sections.**
